<h2>
שימוש בלולאות לסימולציה פיסיקלית בצעדי זמן קטנים
</h2>

אחד השימושים הבולטים בלולאות בהקשר פיסיקלי הוא לביצוע מעקב אחרי ההתפתחות בזמן של מערכת נתונה.
בשיטה זו, מוצבים ערכי התחלה למשתני הבעייה ואז מבוצעת לולאה המעדכנת את המשתנים בהתאם לכלל נתון.

<h3>
תנועה במימד אחד בתאוצה קבועה
</h3>

בדוגמה הבאה מחושב מיקומו של גוף המתחיל ממנוחה ונע בתאוצה קבועה במהלך 10 שניות.  
הריצו את הקוד ולאחר מכן קראו את ההסבר למבנהו.
</div>

```python
t=0
a=10
v=0
x=0

n=10
dt=0.1

for i in arange(n):
    t=t+dt
    x=x+v*dt
    v=v+a*dt

print(t)
print(x)
print(v)
```

<div style="direction:rtl">
כפי שניתן לראות בקוד, ראשית מוזנים ערכים לזמן ההתחלתי לתאוצה, למהירות ולמיקום בהתחלה.
<br>
$$ t=0 $$
$$ a=10 $$
$$ v=0 $$
$$ x=0 $$

שימו לב שאין יחידות. המחשב אינו יודע שהמשתנים מייצגים גדלים פיזיקליים. עבורו אלו הם מספרים ופעולות מתמטיות. האחריות לזכור על אילו יחידות מדובר היא עלינו. 
<br>
בנוסף לתנאי ההתחלה מוגדרים מספר הפעמים שהלולאה תבוצע, n, וגודל הקפיצה בזמן שכל מעבר של הלולאה יחשב, 
 dt.
<br>
כעת מבוצעת הלולאה.
בכל שלב מבוצעות שלוש פעולות:
<br>
עדכון משתנה הזמן לערך החדש
$$ t=t+dt $$
עדכון המיקום לזמן החדש, על בסיס המיקום בזמן הקודם, גודל צעד הזמן והמהירות הנוכחית
$$ x=x+v*dt $$
ועדכון המהירות לזמן החדש, על בסיס המהירות בזמן הקודם, גודל צעד הזמן והתאוצה
$$ v=v+a*dt $$
אנו מתייחסים, אם כן, למהירות ולתאוצה כמהירות ותאוצה קבועות במהלך פרק הזמן הקצר שמשכו dt.
בדוגמא זו התאוצה אכן קבועה, אך המהירות משתנה ככל שמתקדם הזמן, לכן חישוב זה הנו מקורב בלבד.
<br>
תהליך העדכון של המשתנים t, x ו-v מחליף (דורס) את הערך שלהם מצעד הזמן הקודם בערך המחושב עבור צעד הזמן הנוכחי.
שימו לב שמשוואת עדכון המיקום אינה נכונה באופן מדויק עבור תנועה בתאוצה קבועה, למרות זאת היא נכונה בקירוב טוב אם dt קטן מאוד. 
עוד שימו לב לכך שאין שימוש במשתנה הלולאה,i במהלכה. 

<br>
היות ובדוגמא זו התאוצה נותרת קבועה, ניתן היה גם להשתמש בנוסחאות של תנועה בתאוצה קבועה, אך שיטת החישוב בה השתמשנו כאן תתאים גם למקרים של תאוצה שאינה קבועה.

<br>
בסיום קטע הקוד אנו מדפיסים את המיקום הסופי והמהירות הסופית של הגוף.
<h4>
משימה 1
</h4>
שנו את קפיצות הזמן לערכים dt=0.1, 0.5, 1
וראו כיצד התוצאה משתנה.

השוו בין תוצאות אלו לחישוב רגיל ע"י שימוש בנוסחת תנועה בתאוצה קבועה.

שימו לב שהיות וכאשר שינינו את dt, לא שינינו את n, הרי שעבור ערכי dt שונים, הלולאה מסתיימת כאשר ערכי x ו-v מייצגים את המיקום והמהירות בזמנים שונים.
לו היה ברצוננו לחשב את המיקום והמהירות בזמן קבוע, היה עלינו לעדכן, בנוסף ל-dt, גם את n.
</div>


<div style="direction:rtl">

קטע הקוד הבא פותר את הצורך לעדכן את n כאשר משנים את dt.
אנו מגדירים משתנה נוסף, timespan המציין את משך הזמן (בשניות) עבורו ברצוננו לבצע את החישוב.
אם ברצוננו לבצע את החישוב עבור פרק הזמן timespan בצעדי זמן שאורך כל אחד מהם dt, הרי שמספר הפעמים שעל הלולאה להתבצע הוא:
$$ \frac{\text{timespan}}{\text{dt}} $$
ואנו מציבים ערך זה למשתנה n.
הפקודה round  מעגלת שברים לשלמים, ובכך מבטיחה שהלולאה תתבצע על רשימה של מספרים שלמים.
בדוגמא זו אין לכך חשיבות, שכן איננו משתמשים במשתנה הלולאה, i, אך הדבר יידרש בהמשך. 

<h4>
משימה 2
</h4>
כעת, ע"י ניסוי ותהייה, מצאו מהו בקירוב צעד הזמן המקסימלי שעבורו ההפרש בין שתי שיטות החישוב קטן מ-0.01, בזמן 10 שניות.
</div>


```python
a=10
v=0
x=0
t=0

dt=0.1
timespan=10
n=round(timespan/dt)

for i in arange(n):
    t=t+dt
    x=x+v*dt
    v=v+a*dt

print(t)
print(x)
print(v)
```

<div style="direction:rtl">
<h3>
שמירת תוצרי הביניים של החישוב למעקב אחר התפתחות המערכת בזמן
</h3>
בדוגמא הקודמת ביצענו את הלולאה אך לא שמרנו את הערכים לאורך צעדי הזמן השונים.
כך אמנם חישבנו את המיקום והמהירות של הגוף בתום קטע הזמן, אך לא את תוצאות חישובי הביניים.
<br>
בקטע הקוד הבא אנו מבצעים את אותו החישוב בדיוק, אך שומרים את תוצאות החישוב בכל צעד זמן לסדרות העוקבות אחר המיקום והמהירות.
לצורך כך אנו מאתחלים שלוש סדרות: סדרה אחת למעקב אחר הזמן, סדרה אחת למעקב אחר המיקום, וסדרה אחת למעקב אחר המהירות.
<br>
קראו את קטע הקוד הבא, הריצו אותו, ולאחר מכן קראו את ההסבר.
</div>

```python
t=0
a=10
x=0
v=0

dt=0.1
timespan=10
n=round(timespan/dt)

ts=zeros(n)
xs=zeros(n)
vs=zeros(n)

for i in arange(n):
    t=t+dt
    x=x+v*dt
    v=v+a*dt
    ts[i]=t
    xs[i]=x
    vs[i]=v
    
plot(ts,xs,'.')
```

<div style="direction:rtl">
קטע הקוד נפתח בדיוק כמו קטע הקוד הקודם, באתחול המשתנים.
<br>
כעת ברצוננו ליצור שלוש סדרות חדשות, אחת עבור הזמנים, אחת עבור מיקומי הגוף בכל צעד זמן, והשלישית עבור מהירויות הגוף בכל צעד זמן.
על מנת ליצור סדרות באורך n, אנו משתמשים בפקודה zeros.
<br>
לאחר יצירת הסדרות,
הקוד לביצוע בלולאה הוא פשוט.
אנו מעדכנים את ערכי הזמן, המיקום והמהירות של הגוף, כפי שעשינו בדוגמא הקודמת, ושומרים את הערכים החדשים שחושבו למקומות המתאימים בסדרות ts, xs ו-vs.
<br>
לסיום אנו מציירים גרף של מיקום הגוף שחושב כנגד הזמן.

על גבי הגרף מהתא הקודם שמייצג את המיקום בקירוב, הוסיפו קו נוסף מדויק ע״י חישוב אנליטי של המיקום בזמנים השונים בעזרת הנוסחה של מיקום בתאוצה קבועה.

בפרט, ניתן להשתמש באותה השיטה לדמות תנועת גוף גם כאשר התאוצה אינה קבועה אלא תלויה במשתנים האחרים במערכת (הזמן, המיקום, או המהירות, למשל).

שנו שוב את סדר החישוב בגוף הלולאה ובידקו האם וכיצד הוא משפיע.

<h3>
חקירת תנועה בשני מימדים
</h3>
ניתן להשתמש בשיטת ההדמיה שהצגנו ליצירת מודל תנועה של גוף גם ביותר ממימד אחד (2, 3 או יותר מימדים).
כל שעלינו לעשות לשם כך הוא לעדכן משתנים עבור כל מימד.
אם ברצוננו לעקוב אחר המיקום או המהירות בצירים השונים במהלך הסימולציה, עלינו גם לייצר סדרות נוספות בהתאם.
בדוגמא הבאה אנו מדמים זריקת גוף במישור.
אנו עוקבים אחר מהירות הגוף ומיקומו הן בציר האופקי והן בציר האנכי.

```python
t=0
ax=0
ay=-9.8
x=0
vx=7
y=0
vy=20

dt=0.1
timespan=5
n=round(timespan/dt)

ts=zeros(n)
xs=zeros(n)
vxs=zeros(n)
ys=zeros(n)
vys=zeros(n)

for i in arange(0,n):
    t=t+dt
    x=x+vx*dt
    vx=vx+ax*dt
    y=y+vy*dt
    vy=vy+ay*dt
    
    ts[i]=t
    xs[i]=x
    vxs[i]=vx
    ys[i]=y
    vys[i]=vy

plot(xs,ys,'.')   
```
